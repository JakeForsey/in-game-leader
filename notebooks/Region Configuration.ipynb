```python
from PIL import Image

import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("..")
from ingameleader.utils import extract, relative_to, mask_unused_regions, match_signature
from ingameleader import config as cfg
```

```python
def plot_regions_from_full_screen(image_path, normed=True):
    img = Image.open(image_path)
    
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis("off")
    _ = plt.show()

    frame = extract(cfg.META_REGION, np.array(img), (0, 0) + img.size)
    frame = mask_unused_regions(frame)
    
    def _plot_region(region, title):
        plt.title(title)
        plt.imshow(region)
        plt.axis("off")
        _ = plt.show()
    
    _plot_region(
        frame,
        "META_REGION",
    )
    _plot_region(
        extract(cfg.TIME_REGION, frame),
        "TIME_REGION",
    )
    _plot_region(
        extract(cfg.MINUTES_REGION, frame),
        "MINUTES_REGION",
    )
    _plot_region(
        extract(cfg.SECONDS_REGION, frame),
        "SECONDS_REGION",
    )
    _plot_region(
        extract(cfg.MONEY_REGION, frame),
        "MONEY_REGION",
    )
    _plot_region(
        extract(cfg.LOCATION_TEXT_REGION, frame),
        "LOCATION_TEXT_REGION",
    )
    _plot_region(
        extract(cfg.ROUND_TEXT_REGION, frame),
        "ROUND_TEXT_REGION",
    )
    _plot_region(
        extract(cfg.CT_SCORE_REGION, frame),
        "CT_SCORE_REGION",
    )
    _plot_region(
        extract(cfg.T_SCORE_REGION, frame),
        "T_SCORE_REGION",
    )
    return frame
```

```python
def plot_pixel_signature(frame, signature, meta_region=cfg.META_REGION, show_region=None):
    frame = frame.copy()
    for position, colour, expectation in signature:
        position = relative_to(position, meta_region)
        print(frame[position[1], position[0]], np.array(colour), expectation)
        frame[position[1], position[0]] = 255
    
    if show_region is not None:
        frame = extract(show_region, frame, meta_region)
        
    plt.figure(figsize=(20, 10))
    plt.imshow(frame)
    plt.show()
```
